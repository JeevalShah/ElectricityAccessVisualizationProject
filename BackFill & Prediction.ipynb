```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
electricity = pd.read_csv('./Downloads/Electricty_Access_Africa.csv')
electricity.drop(columns=['Country Code'], inplace=True)
```

```python
electricity = electricity.set_index('Country Name')
electricity.head()
```

```python
def fill_missing_values(model, country, country_data):
    
    specific_row = electricity.loc[country]

    missing_years = specific_row.index[specific_row.isnull()].tolist()
    
    for year in missing_years:
        
        X_predict = [[year] for year in missing_years]
        predicted_values = np.clip(model.predict(X_predict), 0, 100)

        # Add the predicted values to the DataFrame
        for year, value in zip(missing_years, predicted_values):
            electricity.at[country, year] = value
```

```python
def train_regression_model(type, country_data):

    country_data = country_data.to_frame().transpose()

    country_data = country_data.dropna(axis=1)

    # If all values are missing, return None
    if country_data.empty:
        return None

    # Select non-empty columns for training
    X_train = country_data.columns
    X_train = X_train.values.reshape(-1, 1)

    # Create a linear regression model
    if type == 1:
        model = LinearRegression()
        print("Linear Regression Model is being used")
    elif type == 2:
        model = Ridge()
        print("Ridge Model is being used")
    elif type == 3:
        model = Lasso()
        print("Lasso Model is being used")
        
    # Fit the model
    model.fit(X_train, country_data.values.flatten())

    return model
```

```python
for country, data in electricity.iterrows():
    #Training Linear Regression
    model = train_regression_model(1, data)

    if model is not None:
        
        fill_missing_values(model, country, data)
        
        # Predict the values for the missing years (2022-2030)
        years_2022_to_2030 = [str(year) for year in range(2022, 2031)]
        X_predict = [[year] for year in range(2022, 2031)]
        predicted_values = np.clip(model.predict(X_predict), 0, 100)

        # Add the predicted values to the DataFrame
        for year, value in zip(years_2022_to_2030, predicted_values):
            electricity.at[country, year] = value
            
electricity = electricity.round(2)
```
