```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/content/df_t.csv")
```

```python

# Set the 'Year' column as the index and convert to datetime
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
df.set_index('Year', inplace=True)
```

```python

# Split your dataset into training and testing sets (80% for training and 20% for testing)
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)
```

```python

# Dictionary to store ARIMA models by country
arima_models = {}
```

```python
# train the ARIMA model and print results
```

```python
from sklearn.metrics import mean_absolute_error
```

```python

# Split data into train and test sets
train_size = int(len(df) * 0.8)
train_data, test_data = df[:train_size], df[train_size:]

# Iterate over columns and train ARIMA models
arima_models = {}
for col in train_data.columns:
    model = ARIMA(train_data[col], order=(1, 1, 1))
    model_fit = model.fit()
    arima_models[col] = model_fit

# Print ARIMA models with column names
for col, model in arima_models.items():
    print(f"ARIMA Model for {col}: {model}")

# Evaluate mean absolute error on test set
mae = {}
for col, model in arima_models.items():
    predictions = model.forecast(steps=len(test_data))
    mae[col] = mean_absolute_error(test_data[col], predictions)

# Save the MAE results (replace 'mae_results.csv' with your desired filename)
mae_df = pd.DataFrame.from_dict(mae, orient="index", columns=["MAE"])
mae_df.to_csv("mae_results.csv")
```
