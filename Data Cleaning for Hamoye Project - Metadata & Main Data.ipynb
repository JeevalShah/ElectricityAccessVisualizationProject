```python
africa = pd.read_csv('./Downloads/Africa_Metadata.csv')
```

```python
africa.drop(columns=['Unnamed: 0'], inplace=True)
africa
```

```python
electricity = pd.read_csv('./Downloads/Electricity Access.csv')
electricity
```

```python
africa_regions = africa['TableName']
africa_regions = africa_regions.tolist()
```

```python
electricity_africa = electricity[electricity['Country Name'].isin(africa_regions)].reset_index(drop=True)
electricity_africa
```

```python
electricity_africa.drop(columns=['Indicator Name', 'Indicator Code', '2022', '2023'], inplace=True)
electricity_africa.set_index('Country Name')
```

```python
electricity_africa_cleaned = electricity_africa.drop(columns=[f'{i}' for i in range(1960, 1990)])
electricity_africa_cleaned
```

```python
for column in range(1990, 2022):
    electricity_africa_cleaned[str(column)] = electricity_africa_cleaned[str(column)].round(2)
```

```python
electricity_africa_cleaned.set_index('Country Name')
```

```python
electricity_africa_cleaned.to_csv('./Downloads/Electricty_Access_Africa.csv')
```

```python
merged_df = pd.merge(africa, electricity_africa_cleaned, on='Country Code', how='right')
merged_df
```

```python
merged_df.drop(columns=['TableName', 'SpecialNotes'], inplace=True)
merged_df.set_index('Country Name')
```

```python
merged_df.to_csv('./Downloads/Meta_Electricty_Access_Africa.csv')
```
